# 3. Image Segmentation & Object Detection

### 1. import

```python
import os
import cv2
import numpy as np
import tensorflow as tf
tf.compat.v1.disable_eager_execution()    # TF1 스타일 그래프 로드용
import urllib.request, tarfile
import matplotlib.pyplot as plt
from ultralytics import YOLO                # YOLOv8
```

### 2. DeepLabv3+ 모델(MobileNetV2 백본, PASCAL VOC 기준) 다운로드 & 로드

```python
MODEL_NAME = "deeplabv3_mnv2_pascal_train_aug"
MODEL_URL  = f"http://download.tensorflow.org/models/{MODEL_NAME}_2018_01_29.tar.gz"

# 모델이 없으면 다운로드하고 압축 해제
model_dir = os.path.join(os.getcwd(), MODEL_NAME)
if not os.path.exists(model_dir):
    os.makedirs(model_dir, exist_ok=True)
    archive_path = os.path.join(model_dir, MODEL_NAME + ".tar.gz")
    print("Downloading DeepLabv3+ model...")
    urllib.request.urlretrieve(MODEL_URL, archive_path)
    print("Extracting...")
    with tarfile.open(archive_path) as tar:
        tar.extractall(path=model_dir)
    os.remove(archive_path)

FROZEN_GRAPH_PATH = os.path.join(model_dir, MODEL_NAME, "frozen_inference_graph.pb")

# DeepLabv3+ 모델 로드 함수
def load_deeplab_model(frozen_graph_path):
    graph = tf.Graph()
    with tf.io.gfile.GFile(frozen_graph_path, "rb") as f:
        graph_def = tf.compat.v1.GraphDef()
        graph_def.ParseFromString(f.read())
    with graph.as_default():
        tf.import_graph_def(graph_def, name="")
    return graph

# DeepLabv3+ 모델 생성
deeplab_graph = load_deeplab_model(FROZEN_GRAPH_PATH)
```

### 3. DeepLabv3+ 추론 & 컬러 매핑

```python
def run_deeplab_inference(graph, image_rgb):
    # image_rgb: H×W×3 numpy array, RGB [0–255]
    with tf.compat.v1.Session(graph=graph) as sess:
        inp = graph.get_tensor_by_name("ImageTensor:0")
        out = graph.get_tensor_by_name("SemanticPredictions:0")
        resized = cv2.resize(image_rgb, (513, 513))
        seg = sess.run(out, feed_dict={inp: [resized]})[0]
        seg = cv2.resize(seg.astype(np.uint8),
                         (image_rgb.shape[1], image_rgb.shape[0]),
                         interpolation=cv2.INTER_NEAREST)
        return seg

# PASCAL VOC color map (21 classes)
_PASCAL_COLORMAP = np.array([
    [  0,   0,   0], [128,   0,   0], [  0, 128,   0], [128, 128,   0],
    [  0,   0, 128], [128,   0, 128], [  0, 128, 128], [128, 128, 128],
    [ 64,   0,   0], [192,   0,   0], [ 64, 128,   0], [192, 128,   0],
    [ 64,   0, 128], [192,   0, 128], [ 64, 128, 128], [192, 128, 128],
    [  0,  64,   0], [128,  64,   0], [  0, 192,   0], [128, 192,   0],
    [  0,  64, 128],
], dtype=np.uint8)

def colorize_segmentation(seg_map):
    # seg_map: H×W array of class IDs [0–20]
    return _PASCAL_COLORMAP[seg_map]
```

### 4. YOLOv8 nano 모델 로드

```python
yolo_model = YOLO('yolov8n.pt')  # Nano 모델 (빠른 테스트용)
```

### 5. 통함 함수

```python
def run_segmentation_and_detection(image_path):
    # 1) 이미지 로드 & RGB 변환
    img_bgr = cv2.imread(image_path)
    img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)

    # 2) DeepLab 세그멘테이션
    seg_map   = run_deeplab_inference(deeplab_graph, img_rgb)
    seg_color = colorize_segmentation(seg_map)
    # 원본 위에 반투명 오버레이
    seg_overlay = cv2.addWeighted(img_rgb, 0.5, seg_color, 0.5, 0)

    # 3) YOLO 추론 (overlay 이미지 기준)
    results = yolo_model(seg_overlay)       # results[0]에 detection 정보
    annotated = results[0].plot()           # 모델 내장 스타일 컬러 사용

    # 4) 결과 출력
    plt.figure(figsize=(12, 6))
    plt.imshow(annotated)
    plt.axis('off')
    plt.title("DeepLabv3+ Segmentation + YOLOv8 Detection")
    plt.show()
```

### 6. 실행

```python
if __name__ == "__main__":
    IMAGE_PATH = "image2.jpg"
    run_segmentation_and_detection(IMAGE_PATH)
```
