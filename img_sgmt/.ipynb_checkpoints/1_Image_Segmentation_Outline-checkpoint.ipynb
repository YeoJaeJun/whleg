# 1. Image Segmentation Outline

### Image Segmentation Models Comparison

| 모델        | 특징                          | 세분화 종류 | 추천 상황                                         |
|-------------|-------------------------------|--------------|--------------------------------------------------|
| U-Net       | 간단, 경계 인식 우수           | Semantic     | 제한된 데이터, 정밀 경계 탐지 필요 시           |
| DeepLabv3+  | 고성능, 다양한 스케일 인식     | Semantic     | 복잡한 환경, 고성능이 요구될 때                |
| Mask R-CNN  | Detection + Segmentation      | Instance     | 객체 인식 + 분할 동시에 필요한 경우              |

### Design Strategy
##### U-Net은 의료용을 목적으로 개발되었다. 그래서 배경이 복잡하고 객체가 많으면 성능이 제한된다.
##### DeepLabv3+는 재난 구조용 로봇에 가장 적합한 모델이지만 객체를 따라 인식하는 게 불가능하다.
##### 그래서 DeepLabv3+와 YOLO를 같이 사용하는 전략을 세운다.

```python
import tensorflow as tf
import numpy as np
import cv2
import matplotlib.pyplot as plt
import os
import tarfile
import tempfile
import urllib.request
```

### 1.1. 사전 학습된 모델 다운로드 및 불러오기(backbone: MobileNetV2)

```python
MODEL_NAME = "deeplabv3_mnv2_pascal_train_aug"
MODEL_URL = f"http://download.tensorflow.org/models/{MODEL_NAME}_2018_01_29.tar.gz"

model_dir = tempfile.mkdtemp()
download_path = os.path.join(model_dir, f"{MODEL_NAME}.tar.gz")

# 모델 다운로드
urllib.request.urlretrieve(MODEL_URL, download_path)

# 압축 해제
tar = tarfile.open(download_path)
tar.extractall(path=model_dir)
tar.close()

frozen_graph_path = os.path.join(model_dir, MODEL_NAME, "frozen_inference_graph.pb")
```

### 1.2. TF 그래프 로드

```python
def load_model(frozen_graph_path):
    graph = tf.Graph()
    with tf.io.gfile.GFile(frozen_graph_path, "rb") as f:
        graph_def = tf.compat.v1.GraphDef()
        graph_def.ParseFromString(f.read())

    with graph.as_default():
        tf.import_graph_def(graph_def, name="")

    return graph

graph = load_model(frozen_graph_path)
```

### 1.3. 이미지 준비 및 전처리 함수

```python
def load_image(path):
    img = cv2.imread(path)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    h, w, _ = img.shape
    resize_ratio = 513.0 / max(h, w)
    target_size = (int(resize_ratio * w), int(resize_ratio * h))
    resized_img = cv2.resize(img, target_size)
    return img, resized_img
```

### 1.4. 추론 함수

```python
def run_inference(graph, image_np):
    with tf.compat.v1.Session(graph=graph) as sess:
        input_tensor = graph.get_tensor_by_name("ImageTensor:0")
        output_tensor = graph.get_tensor_by_name("SemanticPredictions:0")

        batch_seg_map = sess.run(output_tensor, feed_dict={input_tensor: [image_np]})
        seg_map = batch_seg_map[0]
    return seg_map
```

### 1.5. 예측 수행 및 시각화

```python
image_path = 'image2.jpg'
original_img, resized_img = load_image(image_path)

segmentation = run_inference(graph, resized_img)

plt.figure(figsize=(15, 5))
plt.subplot(1, 2, 1)
plt.imshow(original_img)
plt.title("Original Image")
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(segmentation)
plt.title("Segmentation Mask")
plt.axis('off')
plt.show()
```
